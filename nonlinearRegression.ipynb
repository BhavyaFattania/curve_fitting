```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df= pd.read_csv('china_gdp.csv')
df.head()
```

```python
x = df.Year
y = df.Value
plt.scatter(x,y,color = 'blue')
plt.show()
```

```python
Y = y/max(y)
```

```python
X = x/max(x)
```

```python
plt.scatter(X,Y,color = 'blue')
plt.show()
```

```python
def func(X,b1,b2):
    return 1/(1+np.exp(-b1*(X-b2)))
    
```

```python
from scipy.optimize import curve_fit
```

```python
a,b = curve_fit(func,X,Y)
```

```python
b1 = a[0]
b2= a[1]
```

```python
k = func(X,b1,b2)
plt.plot(X,k, color = 'blue')
plt.scatter(X,Y,color = 'red')
plt.show()
```

```python
y_test = Y
y_predict = func(X,b1,b2)
```

```python
import sklearn
from sklearn.metrics import root_mean_squared_error, r2_score
rmse = root_mean_squared_error (y_test,y_predict)

r2 = r2_score(y_test,y_predict)
r2
```
